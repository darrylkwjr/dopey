```python
import pandas as pd
import requests
from django.http import JsonResponse

API_KEY = "wrFFyND6lLARzANikXAxbWjVbWhWHQuM"

def extract_event_data(events):
    event_data = []
    for event in events:
        name = event["name"]
        date = event["dates"]["start"]["localDate"]
        venue = event["_embedded"]["venues"][0]["name"]
        state = event["_embedded"]["venues"][0]["state"]["stateCode"]
        city = event["_embedded"]["venues"][0]["city"]["name"]
        url = event["url"]

        event_data.append({"Name": name, "Date": date, "Venue": venue, "State": state, "City": city, "URL": url})
    return event_data

# Assuming this is part of Django's views.py, modify it to your actual class-based view or function-based view
def get(request, state_code):
    url = "https://app.ticketmaster.com/discovery/v2/events.json"
    params = {
        "apikey": API_KEY,
        "classificationName": "Comedy",
        "size": 40,
        "page": 0,
        "stateCode": state_code
    }

    all_event_data = []

    while True:
        try:
            response = requests.get(url, params=params)
            response.raise_for_status()
            data = response.json()

            if "events" in data["_embedded"]:
                events = data["_embedded"]["events"]
                all_event_data += extract_event_data(events)

        except requests.exceptions.RequestException as e:
            return JsonResponse({"error": f"Unable to fetch data from API. Reason: {e}"}, status=400)

        if "_links" in data and "next" in data["_links"]:
            params["page"] += 1
        else:
            break
    # print(all_event_data)
    print(all_event_data)

    return JsonResponse({"events": all_event_data})

get("request", "TX")
```
